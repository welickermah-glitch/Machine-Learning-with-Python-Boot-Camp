```python
# Problem 1
import numpy as np

prices = np.array([1200, 1250, 1180, 1300, 1220, 1270, 1240])

print('Mean price:' , np.mean(prices).round(2))
print("\n")
print('Highest price:' , np.max(prices))
print("\n")
print('Lowest price:' , np.min(prices))
print("\n")
print('Std Dev:' , np.std(prices).round(2))
# The mean tells the seller about the central valua price of all the prices.
```

```python
# Problem 5
import matplotlib.pyplot as plt

attendance = [35, 37, 40, 38, 42, 39, 41]
days = range(1, len(attendance) + 1)

# Create line graph with markers
plt.plot(days, attendance, marker='o')

# Add title and axis labels
plt.title('School Attendance')
plt.xlabel('Day')
plt.ylabel('Number of Students')

plt.show()
```
